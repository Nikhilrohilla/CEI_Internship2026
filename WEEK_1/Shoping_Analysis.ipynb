```python
numerical_cols=csv_x.select_dtypes(include=['int64','float64']).columns
for col in numerical_cols:
  csv_x[col]=csv_x[col].fillna(csv_x[col].mean())
```

```python
categorical_cols=csv_x.select_dtypes(include=['object']).columns
for col in categorical_cols:
  csv_x[col]=csv_x[col].fillna(csv_x[col].mode()[0])
```

```python
csv_x.isnull().sum()
```

```python
print(csv_x.columns.tolist())
```

```python
print(csv_x['final_price'].dtype)
print(csv_x['final_price'].head())
```

```python
csv_x['final_price']=(
    csv_x['final_price']
    .astype(str)
    .str.replace('"','',regex=False)
    .str.replace('₹','',regex=False)
    .str.replace(',','',regex=False)
)
csv_x['final_price']=pd.to_numeric(csv_x['final_price'],errors='coerce')
print(csv_x['final_price'].head())
print(csv_x['final_price'].dtype)
```

```python
filtered_data=csv_x[csv_x['final_price']>1500]
filtered_data.head()
```

```python
print("Before Duplicate",csv_x.shape)
csv_x.drop_duplicates(inplace=True)
print("After Duplicate",csv_x.shape)
```

```python
csv_x['quantity']=1.5
csv_x['total_amount']=csv_x['final_price']*csv_x['quantity']
csv_x[['title','final_price','quantity','total_amount']].head()
```

```python
csv_x.to_csv('cleaned_dataset.csv',index=False)
print("Cleaned dataset saved successfully!")
```
